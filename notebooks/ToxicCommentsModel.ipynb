```
#!fsharp
type ISentimentRow =
    abstract member comment_text: string
    abstract member toxic: float

[<CLIMutable>]
type SentimentData =
    { Text: string;
      [<ColumnName("Label")>]Toxic: bool }

    static member ForText text =
        { Text = text; Toxic = false}

[<CLIMutable>]
type SentimentPrediction =
    { [<ColumnName("PredictedLabel")>]Prediction: bool
      Probability: float32
      Score: float32 }

let data =
    tidyMinimalistic.GetRowsAs<ISentimentRow>()
    |> Series.values
    |> Seq.map (fun sentimentRow -> { Text = sentimentRow.comment_text; Toxic = if sentimentRow.toxic > 0. then true else false })
```

```
#!fsharp
let context = new MLContext(seed = 1)
```

```
#!fsharp
let dataView = context.Data.LoadFromEnumerable(data)
let split = context.Data.TrainTestSplit(dataView, testFraction = 0.2)
```

```
#!fsharp
let estimator =
    context.Transforms.Text.FeaturizeText(outputColumnName = "Features", inputColumnName = "Text")
        .Append(context.BinaryClassification.Trainers.SdcaLogisticRegression(labelColumnName = "Label", featureColumnName = "Features"))
```

```
#!fsharp
let model = estimator.Fit split.TrainSet
```

```
#!fsharp
let predictions = model.Transform(split.TestSet)
let metrics = context.BinaryClassification.Evaluate(predictions, "Label")
```

```
#!fsharp
sprintf "Accuracy: %f" metrics.Accuracy |> display
sprintf "Area under Curve: %f" metrics.AreaUnderRocCurve |> display
sprintf "F1 Score: %f" metrics.F1Score |> display
```

```
#!fsharp
let predictionFunction = context.Model.CreatePredictionEngine<SentimentData, SentimentPrediction>(model)
```

```
#!fsharp
predictionFunction.Predict(SentimentData.ForText "that is really great")
```
